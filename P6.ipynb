Question-Answering

```python
from transformers import pipeline
```

```python
qa = pipeline("question-answering", model="distilbert/distilbert-base-cased-distilled-squad")
```

```python
context = r'''
Extractive Question Answering is a task to extract an answer from a text given in a question
An example Questioning Answering dataset is SQuAD Dataset, Which is essentially
based on task.If you would like to finetune model on SQuAD task,you may
leverage the examples/pytorch/question/-answering/run_squad.py script
'''
```

```python
question = '''\
What is a good example of Question Answering Dataset?
'''
```

```python
result = qa(question=question,context=context)
```

```python
print(result)
```

```python
c = r'''
Quantum computing utilizes qubits, which unlike classical bits, can exist in multiple states simultaneously due to quantum principles like superposition and entanglement. This allows quantum computers to perform certain calculations exponentially faster than classical computers. Despite its promise, quantum computing is still in early development stages, facing challenges before widespread practical use.
'''
```

```python
q = '''\
What is quantum computing?
'''
```

```python
r1 = qa (q,c)
```

```python
print(r1)
```
