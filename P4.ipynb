**Sentence Similarity**

```python
#sentence similarity measures how close two sentences are.
# Sentence similarity is useful in information retrieval and clustering/grouping information.
```

```python
!pip install sentence-transformers
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Sentence similarity models convert input text into embeddings(vectors).
# This embeddings captures semantic informations
```

```python
s1 = ['my name is om','Myself om','Hey i am om','Om this side']
```

```python
embeddings = model.encode(s1, convert_to_tensor = True)
```

```python
# Using `convert_to_tensor=True` ensures that input data is automatically converted
# to the tensor format required by the machine learning framework. This promotes consistency,
#  compatibility with tensor-specific operations, and potentially better performance.
#  However, it may incur overhead and alter original data, which could be undesirable
#  in certain cases where flexibility and data integrity are prioritized.
#  Therefore, its usage depends on balancing these factors according to the specific needs of the application.
```

```python
embeddings
```

```python
s2 = ['My name is Krishna','Myself Krishna Patel','hey,Krishna This side!!']
```

```python
em2 = model.encode(s2, convert_to_tensor = True)
```

```python
em2
```

```python
from sentence_transformers import util
```

```python
#we use cosine methods from util to check semantic similarity
```

```python
cosine_scores = util.cos_sim(embeddings,em2)
```

```python
#Pair wise similarity of every sentence in list s1 to every sentence in s2
print(cosine_scores)
```

```python
#scoring compared to each sentence in list s2 with list s1
for i in range (len(s2)):
  print("{} \t\t {} \t\t Score: {:.2f}".format(s2[i],s1[i],cosine_scores[i][i]))
```
