Building with gradio

```python
import gradio as g
demo = g.Blocks()
```

```python
def transcribe(filepath):
  if filepath is None:
    g.Warning('no audio found, please try again')
    return ""
  output = asr(filepath)
  return output['text']
```

```python
#creating UI for importing audio file , and clicking on a button to generate text

mic = g.Interface(
    fn = transcribe,
    inputs = g.Audio(sources='microphone',type='filepath'),
    outputs = g.Textbox(label="Transcription",lines=3),
    allow_flagging='never'
)
```

```python
file_upload = g.Interface(
    fn = transcribe,
    inputs = g.Audio(sources='upload',type='filepath'),
    outputs = g.Textbox(label="Transcription",lines=3),
    allow_flagging='never'
)
```

```python
with demo:
  g.TabbedInterface(
      [mic,file_upload],
      ["Transcibe Microphone","Transcribe Audio File"],
      )

# demo.launch(debug=True,share=True)
demo.launch(debug=True)
```
