Text Classification

```python
from transformers import pipeline,Conversation
```

```python
#sentiment Analysis
classifier = pipeline('text-classification',model="lxyuan/distilbert-base-multilingual-cased-sentiments-student")
```

```python
t1 = 'Indian Government focuses on Economic Growth'
```

```python
classifier(t1)
```

```python
t2 = 'India is eradicating extreme poverty'
classifier(t2)
```

```python
t3 = 'Wester Media is hateful towards india'
classifier(t3)
```

```python
t4 = 'sun rises from east'
classifier(t4)
```

Zero Shot Classfication

```python
zero = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

```

```python
s1 = "India is the fastest rising economy"
labels = ['Country','Finance','Economics','Food']
```

```python
c = zero(s1,labels,multi_label=True)
```

```python
print(c)
```

```python
print(c['sequence'])
print(c['labels'])
print(c['scores'])
```
